### Support Vector Machine

#### Support vector machine draws a hyper plane in n dimensional space such that is maximizes margin between classification groups.

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
```

```python
df.head()
```

```python
dir(iris)
```

```python
iris.target
```

```python
# Using a column value to generate values for the next or new column using pandas.series.apply(lambda expression) feature
df['flower_name'] = df.target.apply(lambda x : iris.target_names[x])
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]

plt.figure(figsize = (10, 6), dpi = 100)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'green', marker = '+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'blue', marker = '.')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color = 'orange', marker = 'v')
plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color = 'green', marker = '+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color = 'blue', marker = '.')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color = 'orange', marker = 'v')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['target', 'flower_name'], axis = 'columns')
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
```

```python
dir(model)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.datasets import load_digits
img_data = load_digits()
```

```python
dir(img_data)
```

```python
plt.imshow(img_data.images[3])
```

```python
new_df = pd.DataFrame(img_data.data, columns = img_data.feature_names)
```

```python
X = new_df.copy()
y = img_data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='poly', C=10, gamma = 100)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
