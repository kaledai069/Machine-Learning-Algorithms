There is so much option for the parameters to choose from. Hyperparameter tuning is the technique for optimally choosing those parameter.

```python
import pandas as pd
from sklearn.datasets import load_iris
iris_data = load_iris()
df = pd.DataFrame(iris_data.data, columns = iris_data.feature_names)
df['flower'] = iris_data.target
df['flower'] = df['flower'].apply(lambda x : iris_data.target_names[x])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('flower', axis = 'columns'), iris_data.target)
```

```python
from sklearn.svm import SVC
model = SVC(kernel = 'rbf', C = 30, gamma = 'auto')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cv_val = 12
cross_val_score(SVC(kernel = 'rbf', C = 30, gamma = 'auto'), iris_data.data, iris_data.target, cv = cv_val).mean()
```

```python
cross_val_score(SVC(kernel = 'linear', C = 30, gamma = 'auto'), iris_data.data, iris_data.target, cv = cv_val).mean()
```

```python
cross_val_score(SVC(kernel = 'poly', C = 30, gamma = 'auto'), iris_data.data, iris_data.target, cv = cv_val).mean()
```

```python
kernels = ['rbf', 'linear', 'poly', 'sigmoid']
c = range(1, 30)
cv = range(2, 10)
avg_scores = {}
for kval in kernels:
    for cval in c:
        for cv_val in cv:
           avg_scores[kval + '_' + str(cval) + '_' + str(cv_val)] = cross_val_score(SVC(kernel = str(kval), C = cval, gamma = 'auto'), iris_data.data, iris_data.target, cv = cv_val).mean()
       
optimal_value = max(avg_scores, key = avg_scores.get)
optimal_value
```

```python
cross_val_score(SVC(kernel = 'linear', C = 1, gamma = 'auto'), iris_data.data, iris_data.target, cv = 3)
```
