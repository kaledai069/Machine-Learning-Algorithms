Random Forest Algorithm is one of the bagging technique but here features are also randomly sampled.

```python
import pandas as pd
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.describe()
```

```python
# checking target imbalance state

df.Outcome.value_counts()
```

```python
X = df.drop('Outcome', axis = 'columns')
y = df.Outcome
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, startify = y, test_size = 0.2)
```
