## Univariate Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')

plt.plot(df['area'].tolist(), df['price'].tolist())
plt.show()
```

```python
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.xlabel('area (sqr ft)')
plt.ylabel('price (US$)')
plt.show()
```

```python
# As simple as it gets here
reg = linear_model.LinearRegression().fit(df[['area']], df.price)
reg.predict([[3300]])
```

```python
# slope part or the theta1 from Andrew Ng
reg.coef_
```

```python
# intercept or the theta0 (base value) from Andrew Ng
reg.intercept_
```

```python
areas_values = pd.read_csv('areas.csv')
new_list_value = list()
for value in areas_values.area:
    new_list_value.append(reg.predict([[value]])[0])
plt.plot(areas_values.area, new_list_value)
plt.show()
```

```python
new_df = pd.read_csv('areas.csv')
p = reg.predict(new_df) # Noice and easy way to predict for whole dataframe
new_df['prices'] = p
new_df.to_csv('predicted_values.csv', index = False)
```

```python
# Plotting the inital data and the regressionlines
df = pd.read_csv('homeprices.csv')
plt.figure(figsize = (8, 6))
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.plot(new_df.area, p, 'b-')
```

```python
# Some Exercise
df = pd.read_csv('Exercise/canada_per_capita_income.csv')
plt.scatter(df.year, df['per capita income (US$)'])
```

```python
reg_per_capita = linear_model.LinearRegression().fit(df[['year']], df['per capita income (US$)'])
```

```python
reg_per_capita.predict([[2020]])
```

```python
plt.scatter(df.year, df['per capita income (US$)'], color = 'orange', marker = 'x', label = 'Original Data')
plt.plot(df.year, reg_per_capita.predict(df[['year']]), 'b-', label = 'Predicted Data')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US $)')
plt.legend()
plt.savefig('predicted_data.png', dpi = 240)
plt.show()
```
