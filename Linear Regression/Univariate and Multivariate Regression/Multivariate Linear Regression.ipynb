## Multivariate Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices_.csv')
df.head(5)
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
```

```python
multi_reg = linear_model.LinearRegression().fit(df[['area', 'bedrooms', 'age']].values, df.price.values)
```

```python
multi_reg.coef_
```

```python
multi_reg.predict([[3000, 3, 50]])
```

```python
# Exercise from this section

from word2number import w2n
hire_df = pd.read_csv('Exercise/hiring.csv')
hire_df.experience = hire_df.experience.fillna('one')
hire_df['test_score(out of 10)'] = hire_df['test_score(out of 10)'].fillna(hire_df['test_score(out of 10)'].median())
hire_df
```

```python
for index, value in hire_df.experience.items():
    hire_df.experience = hire_df.experience.replace([value], w2n.word_to_num(value))

```

```python
hire_multi_reg = linear_model.LinearRegression().fit(hire_df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']].values, hire_df['salary($)'].values)
```

```python
hire_multi_reg.predict([[2, 9 , 6]])
```

```python
hire_multi_reg.predict([[12, 10, 10]])
```
